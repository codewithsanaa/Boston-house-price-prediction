```python
#split the data into training and testing 
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Divide the data into dependent and independent variables
X = boston_data.iloc[:,:-1]
Y = boston_data.iloc[:,-1]
Y
X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
x_train.shape;x_test.shape
```

```python
#building the model
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
#Create a table showing the coefficient (influence)
#of each of the columns
cdf = pd.DataFrame(lin_reg.coef_,
                  index=x_train.columns
                ,
                  columns=['Coefficients'])
cdf
```

```python
#feature scaling is to be done to make every column standard
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#building the model
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
x_train
```

```python
#Create a table showing the coefficient (influence)
#of each of the columns
cdf = pd.DataFrame(lin_reg.coef_,
                   index=d.feature_names,columns=['Coefficients'])
cdf
```

```python
#then we can go fro prediction
reg_predict = lin_reg.predict(x_test)
```

```python
#R-squared value - coefficient of determination
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,reg_predict))
print(r2_score(y_test,reg_predict)*100)
```

```python
#we can go for predictions by passing our inputs
X.iloc[0];Y[0]
```

```python
#lets check what our model has given
lin_reg.predict(X.iloc[[0]])
```

```python
#For sure target will be negative as feature scaling to be done
#as you can observe price is negative we need to scale it
scaled = lin_reg.predict(scaler.transform(X.iloc[[0]]))
scaled
```

```python
#once we are confident enough to test the model in production - ready environment
#we save the models using pickle
import pickle
```

```python
#we will create pickle files for both regressionmodel and scaling model
#open() is generally for file modes
pickle.dump(lin_reg,open('regmodel.pkl','wb'))
```

```python
#some way we create another pickle file for scaling model
pickle.dump(scaler,open('scaler.pkl','wb'))
```

```python
#now lets give some validation data and test on pickle files
pickled_model = pickle.load(open('regmodel.pkl','rb'))
```

```python
#do some for transformation
f = pickle.load(open('scaler.pkl','rb'))
```

```python
#as we have successfully loaded the picklefiles we pass some data
pickled_model.predict(f.transform(X.iloc[[0]]))
```
